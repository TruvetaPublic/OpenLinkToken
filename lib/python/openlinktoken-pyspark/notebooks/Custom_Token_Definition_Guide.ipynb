## Method 1: Quick Token from Exchange Config (Recommended)

Create a custom T6 token in one line with the rule:
`U(last-name)|U(first-name)|birth-date|postal-code-3|U(sex)`


```python
# Create a T6 token generator in one call
generator = quick_token_from_exchange_config(
    "T6",
    [
        ("last_name", "T|U"),
        ("first_name", "T|U"),
        ("birth_date", "T|D"),
        ("postal_code", "T|S(0,3)"),
        ("sex", "T|U"),
    ],
    exchange_config_path=EXCHANGE_CONFIG_PATH,
    private_key_path=PRIVATE_KEY_PATH,
    private_key_env=PRIVATE_KEY_ENV,
)

# Test it with sample data
person_attrs = {
    RecordIdAttribute: "1",
    FirstNameAttribute: "John",
    LastNameAttribute: "Doe",
    BirthDateAttribute: "1990-01-15",
    SexAttribute: "Male",
    PostalCodeAttribute: "98101",
}

result = generator.get_all_tokens(person_attrs)
print(f"T6 Token: {result.tokens.get('T6')}")
```

## Method 2: Token Builder (More Flexible)

Use the fluent TokenBuilder API for more control:

```python
# Create a custom T6 token
t6_token = (
    TokenBuilder("T6")
    .add("last_name", "T|U")
    .add("first_name", "T|U")
    .add("birth_date", "T|D")
    .add("postal_code", "T|S(0,3)")
    .add("sex", "T|U")
    .build()
)

# Create a custom token definition
custom_definition = CustomTokenDefinition().add_token(t6_token)

# Create token generator
generator = create_token_generator_from_exchange_config(
    exchange_config_path=EXCHANGE_CONFIG_PATH,
    private_key_path=PRIVATE_KEY_PATH,
    private_key_env=PRIVATE_KEY_ENV,
    token_definition=custom_definition,
)

# Generate tokens
result = generator.get_all_tokens(person_attrs)
print(f"T6 Token: {result.tokens.get('T6')}")
```

## Method 3: Multiple Custom Tokens

Define multiple tokens in one definition:

```python
# Create T6 token: U(last-name)|U(first-name)|birth-date|postal-code-3|U(sex)
t6_token = (
    TokenBuilder("T6")
    .add("last_name", "T|U")
    .add("first_name", "T|U")
    .add("birth_date", "T|D")
    .add("postal_code", "T|S(0,3)")
    .add("sex", "T|U")
    .build()
)

# Create T7 token: U(last-name-3)|U(first-name-3)|birth-date
t7_token = (
    TokenBuilder("T7").add("last_name", "T|S(0,3)|U").add("first_name", "T|S(0,3)|U").add("birth_date", "T|D").build()
)

# Add both tokens to definition
custom_definition = CustomTokenDefinition().add_token(t6_token).add_token(t7_token)

# Create generator
generator = create_token_generator_from_exchange_config(
    exchange_config_path=EXCHANGE_CONFIG_PATH,
    private_key_path=PRIVATE_KEY_PATH,
    private_key_env=PRIVATE_KEY_ENV,
    token_definition=custom_definition,
)

# Generate both tokens
result = generator.get_all_tokens(person_attrs)
print(f"T6 Token: {result.tokens.get('T6')}")
print(f"T7 Token: {result.tokens.get('T7')}")
```

## Using Custom Tokens with PySpark DataFrames

The recommended production pattern is to pass your custom `TokenDefinition` to `OpenLinkTokenProcessor.from_exchange_config(...)`.


```python
from pyspark.sql import SparkSession

from openlinktoken_pyspark import CustomTokenDefinition, TokenBuilder

# Create Spark session
spark = SparkSession.builder.appName("CustomTokens").config("spark.sql.shuffle.partitions", "4").getOrCreate()

# Create sample DataFrame with ALL required columns (even if not used by custom token)
# Note: Currently OpenLinkTokenProcessor validates ALL standard columns regardless of custom tokens
data = [
    ("1", "John", "Doe", "1990-01-15", "Male", "98101", "123-45-6789"),
    ("2", "Jane", "Smith", "1985-06-20", "Female", "94105", "987-65-4321"),
]
df = spark.createDataFrame(
    data, ["RecordId", "FirstName", "LastName", "BirthDate", "Sex", "PostalCode", "SocialSecurityNumber"]
)

# Step 1: Define your custom T6 token
t6_token = (
    TokenBuilder("T6")
    .add("last_name", "T|U")
    .add("first_name", "T|U")
    .add("birth_date", "T|D")
    .add("postal_code", "T|S(0,3)")
    .add("sex", "T|U")
    .build()
)

# Step 2: Create custom token definition
custom_definition = CustomTokenDefinition().add_token(t6_token)

# Step 3: Create processor with custom definition
processor = OpenLinkTokenProcessor.from_exchange_config(
    exchange_config_path=EXCHANGE_CONFIG_PATH,
    private_key_path=PRIVATE_KEY_PATH,
    private_key_env=PRIVATE_KEY_ENV,
    token_definition=custom_definition,
)

# Step 4: Process DataFrame with custom tokens
tokens_df = processor.process_dataframe(df)

# Show results - you'll see T6 tokens instead of T1-T5!
print("Custom T6 tokens generated:")
print("DataFrame structure:")
tokens_df.printSchema()
tokens_df.show(truncate=False)
```

### Multiple Custom Tokens with PySpark

You can also use multiple custom tokens with the same exchange-config flow:


```python
# Define multiple custom tokens
t6_token = (
    TokenBuilder("T6")
    .add("last_name", "T|U")
    .add("first_name", "T|U")
    .add("birth_date", "T|D")
    .add("postal_code", "T|S(0,3)")
    .add("sex", "T|U")
    .build()
)

t7_token = (
    TokenBuilder("T7").add("last_name", "T|S(0,3)|U").add("first_name", "T|S(0,3)|U").add("birth_date", "T|D").build()
)

# Add both to definition
multi_definition = CustomTokenDefinition().add_token(t6_token).add_token(t7_token)

# Create processor with multiple custom tokens
processor_multi = OpenLinkTokenProcessor.from_exchange_config(
    exchange_config_path=EXCHANGE_CONFIG_PATH,
    private_key_path=PRIVATE_KEY_PATH,
    private_key_env=PRIVATE_KEY_ENV,
    token_definition=multi_definition,
)

# Process - will generate both T6 and T7 tokens
multi_tokens_df = processor_multi.process_dataframe(df)

print("Multiple custom tokens (T6 and T7):")
multi_tokens_df.show(truncate=False)
```

## Token Processing Options

For production encrypted tokens, prefer `OpenLinkTokenProcessor.from_exchange_config(...)` so exchange-config inputs are resolved on the driver.

`OpenLinkTokenProcessor` also supports three direct-constructor modes for backward compatibility and experimentation:

1. **Hash + Encrypt** (Compatibility): Explicit raw secrets for existing integrations
2. **Hash Only** (Experimentation): Faster, tokens are hashed but not encrypted  
3. **Plain Text** (Debugging): Raw concatenated strings like `DOE|JOHN|1990-01-15|981|MALE`

Simply omit the `hashing_secret` and/or `encryption_key` to control the direct-constructor mode:

## Summary

Three ways to create custom tokens:

1. **`quick_token_from_exchange_config()`** - Fastest recommended one-liner for exchange-config workflows
2. **`TokenBuilder` + `create_token_generator_from_exchange_config()`** - Fluent API for readable, flexible definitions
3. **Direct constructors / helpers** - Backward-compatible raw-secret control for existing integrations

Choose the method that best fits your workflow!
